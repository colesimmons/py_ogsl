```python
dicts = []
for sign in ogsl.signs:
    sign_dict = sign.model_dump(exclude={"values", "forms"})
    sign_dict["type"] = "Sign"
    dicts.append(sign_dict)

    for value in sign.values:
        value_dict = value.model_dump()
        value_dict["type"] = "SignValue"
        dicts.append(value_dict)
    
    for form in sign.forms:
        form_dict = form.model_dump(exclude={"values"})
        form_dict["type"] = "Form"
        dicts.append(form_dict)
        
        for value in form.values:
            value_dict = value.model_dump()
            value_dict["type"] = "FormValue"
            dicts.append(value_dict)
dicts
```

```python
import pandas as pd
df = pd.DataFrame(dicts)
df = df.fillna('')
df
```

```python
import json
def value_to_signs_json():
    out = {}
    mult_forms = set()
    
    def add(key, val):
        if key in out:
            mult_forms.add(key)
        out[key] = val

    # ----------
    # VALUES
    # ----------

    # First pass:
    # Sometimes values are present for both a sign and a form.
    # We want to distinguish when that actually represents multiple
    # cuneiform forms (i.e. different glyphs) or when a "form" is
    # really just a different name for the same sign (e.g. KID and E2)
    for sign in ogsl.signs:
        # Add sign values -> sign cuneiform
        if sign.unicode_cuneiform:
            for value in sign.values:
                add(value.value, sign.unicode_cuneiform)

        # Add form values -> form cuneiform
        for form in sign.forms:
            if form.unicode_cuneiform:
                for value in form.values:
                    add(value.value, form.unicode_cuneiform)

    # Second pass:
    # Now we only deal with values that didn't have a unicode glyph
    # associated with their respective sign or form.
    # Here we prefer the simple sign name, and only add forms if they are compound.
    for sign in ogsl.signs:
        # Add sign values -> sign name
        for value in sign.values:
            if value.value not in out:
                add(value.value, sign.name)

        # Add form values -> form name
        for form in sign.forms:
            for value in form.values:
                if value.value not in out and "|" in form.variant_code:
                    add(value.value, form.variant_code)
    
    # ----------
    # OTHER
    # ----------
    for sign in ogsl.signs:
        # Add sign name -> sign cuneiform
        if sign.unicode_cuneiform:
            add(sign.name, sign.unicode_cuneiform)

        # Add sign aka
        if sign.aka and sign.unicode_cuneiform:
            add(sign.aka, sign.unicode_cuneiform)
            
        for form in sign.forms:
            # Add form name -> form cuneiform
            if form.unicode_cuneiform:
                add(form.variant_code, form.unicode_cuneiform)

            # Add form aka
            if form.aka and form.unicode_cuneiform:
                add(form.aka, form.unicode_cuneiform)
            

    print(f"Of {len(out.keys())} values, {len(list(mult_forms))} have multiple forms")
    for key in mult_forms:
        del out[key]

    with open("signs-3.json", "w") as outfile:
        json.dump(out, outfile, ensure_ascii=False)

value_to_signs_json()
```
