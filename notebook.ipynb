```python
from parse import get_blocks, limit_to_sign_blocks, parse_sign_blocks
```

```python
blocks = get_blocks('ogsl.asl')
sign_blocks = limit_to_sign_blocks(blocks)
signs = parse_sign_blocks(sign_blocks)
```

```python
form_dicts = []
for sign in signs:
    for form in sign.forms:
        form_dicts.append(form.model_dump())
form_dicts
```

```python
import pandas as pd
form_df = pd.DataFrame(form_dicts)
form_df
```
